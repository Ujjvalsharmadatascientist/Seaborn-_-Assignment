```python
import seaborn as sns
fmri =sns.load_dataset('fmri')
fmri.head()
```

```python
fmri['event'].value_counts()
```

```python
fmri['region'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
sns.lineplot(data=fmri,x='timepoint',y='signal',style='region',hue='event') 
plt.title('Signal over Time for Different Events and Regions')
plt.show()
```

```python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
sns.histplot(data=diamonds, x='price',hue='cut',kde=True)
plt.title('Distribution of Price for various diamond cut')
plt.show()
```
